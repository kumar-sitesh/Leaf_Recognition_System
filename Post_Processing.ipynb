```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
result = pd.read_csv('result.csv',index_col=0)
result = result.T
result.columns = "AR WR PtoA PHtP AHTAL".split()
result.head()
```

```python
hist = pd.read_csv('histo.csv',index_col=0)
```

```python
column = []
for i in range(0,768):
    column.append('c'+str(i))
    
hist.columns=column
```

```python
hist.head()
```

```python
hog = pd.read_csv('hog_gray.csv',index_col=0)
cols = []
for i in range(len(hog.columns)):
    cols.append('hog'+str(i))
hog.columns=cols
```

```python
hist['AR']=np.array(result['AR'])
hist['WR']=np.array(result['WR'])
hist['PtoA']=np.array(result['PtoA'])
hist['PHtP']=np.array(result['PHtP'])
hist['AHTAL']=np.array(result['AHTAL'])
for i in hog.columns:
    hist[i] = hog[i]
```

```python
hist
```

```python
index=hist.index
```

```python
df=hist
```

```python
hist.to_csv('final_result.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,index,
                                                    test_size=0.25,random_state=401)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

# Will take some time
for i in range(1,20):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,20),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.savefig('error_vs_k.png')
```
